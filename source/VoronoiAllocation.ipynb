I use a k-means clustering to asign galaxies to different Voronoi cells

```python
n_points =size(observed_gals.x)
features = empty((n_points,2))
features[:,0] = observed_gals.x[:]
features[:,1] = observed_gals.y[:]
```

```python
n_clusters = 4500
centroids,_ = kmeans(features,n_clusters, iter=1)
idx,_ = vq(features,centroids)
n_clusters = size(centroids[:,0])
```

```python
#scatter(features[:,0], features[:,1], s=0.1)
scatter(centroids[:,0],centroids[:,1], color='red',s=10,alpha=0.4)
```

Count the number of galaxies in each group and its approximate size

```python
MIN_LEN = 1.0E-3
cluster_size = empty((0))
sigma_x = empty((0))
sigma_y = empty((0))
for i in range(n_clusters):
    id_group = where(idx==i)
    id_group = id_group[0]
    tmp_x = features[id_group,0]
    tmp_y = features[id_group,1]
    if(size(id_group)<=2):
        sigma_x = append(sigma_x, MIN_LEN)
        sigma_y = append(sigma_y, MIN_LEN)
    else:
        sigma_x = append(sigma_x, std(tmp_x))
        sigma_y = append(sigma_y, std(tmp_y))
    cluster_size = append(cluster_size, size(id_group))
    
```

```python
area_norm = (sigma_x * sigma_y)/(sum(sigma_x*sigma_y))
priority = log10(1.0/area_norm)
scatter(cluster_size, priority)
```
