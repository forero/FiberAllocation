```python
#Poisson - BigBoss
ra_range = 3.0
fiber_pitch_mm = 7.56
fiber_size_mm = 3.6
patrol_radius_mm = 7.0
plate_scale = ra_range/75.0/fiber_pitch_mm
method="galaxy_density"
gal_density  = 2.0*(119.0*137.0/(ra_range**2))
mock_gals = AM.MockGalaxyCatalog(number_density=gal_density, ra_range = ra_range, dec_range=ra_range, random=True)
mock_gals.tile_ID[:] = -1
for i in range(5):    
    BB_fibers = AM.FiberSet(Nx=75, Ny=87,fiber_pitch=fiber_pitch_mm * plate_scale, fiber_size=fiber_size_mm * plate_scale, rectangle=True)
    AM.make_fiber_allocation(BB_fibers, mock_gals, visit_ID=i, rank_criterion=method, patrol_radius=patrol_radius_mm * plate_scale,\
                    exclusion_radius=BB_fibers.fiber_size * 0.5)
    allocated_gals = where(mock_gals.visit_ID!=-2)
    n_allocated = size(allocated_gals)
    n_total_gals = 1.0*mock_gals.Ngalaxies
    print 'fraction of allocated galaxies', n_allocated/n_total_gals, n_total_gals, BB_fibers.Nfibers
scatter(BB_fibers.x, BB_fibers.y, color='blue', alpha=1.0, s=1.0)
```
